```python
emo = pd.read_csv('./Emoji_Codes.csv')
emo_a = pd.np.array(emo)
```

```python
emo_dict = {('[EMO, %d]' % item[1]): item for item in emo_a}
```

```python
p = re.compile('\[EMO, \d+\]')
```

```python
counts = defaultdict(int)
```

```python
for text in  df['Message']:
	matches = re.findall(p, text)
	if matches:
		for match in matches:
			if match in emo_dict:
				matched_emo = emo_dict[match][0]
				counts[matched_emo] += 1
```

```python
sorted_counts = sorted(counts.items(), key=lambda x: x[1])
```

By checking the table of Emoji, the Emoji Order: [EMO,8],[EMO,23],[EMO,7],[EMO,2],[EMO,15]

```python
def transfer(x):
    Temp_list = []
    for item in df['Message']:
        if ('[EMO, %d]' %x) in item:
            Temp_list.append(item)
    Temp_list = [re.sub(p, '', item) for item in Temp_list]
    Temp_list = np.array([x for x in Temp_list if x != ''])
    Temp_df = pd.DataFrame(Temp_list)
    Temp_df.columns = ['text']
    a = np.empty(len(Temp_list))
    a.fill(x)
    Temp_df['Label'] = a
    return(Temp_df)
```

```python
Data = pd.DataFrame(columns=['text','Label'])
```

```python
Data = Data.append(transfer(8), ignore_index = True)
```

```python
Data = Data.append(transfer(23), ignore_index = True)
```

```python
Data = Data.append(transfer(7), ignore_index = True)
```

```python
Data = Data.append(transfer(2), ignore_index = True)
```

```python
Data = Data.append(transfer(15), ignore_index = True)
```

```python
Data
```

# Define train&test data

```python
y=Data['Label'].values
X=Data['text'].values
```

```python
from sklearn import cross_validation
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.3, random_state=0)
```

# Vectorize data

We use count vectorizer and tfidf vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# unigram_count = CountVectorizer(encoding='latin-1', min_df=5, stop_words=None)
```

```python
unigram_tfidf = TfidfVectorizer(encoding='latin-1', min_df=5, use_idf=True, smooth_idf=True, norm=u'l2')
```

```python
X_train_vec = unigram_tfidf.fit_transform(X_train)
```

# Naive Bayes Classification

```python
from sklearn.naive_bayes import MultinomialNB
nb_clf= MultinomialNB()
nb_clf.fit(X_train_vec,y_train)
```

```python
X_test_vec = unigram_tfidf.transform(X_test)
```

```python
nb_clf.score(X_test_vec,y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = nb_clf.fit(X_train_vec, y_train).predict(X_test_vec)
cm = confusion_matrix(y_test, y_pred)
print (cm)
```

```python
from sklearn.metrics import precision_score
precision_score(y_test, y_pred, average=None)
```

# LR Classification

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train_vec,y_train)
```

```python
lr.score(X_test_vec,y_test)
```
